```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pickle
```

```python
df=pd.read_csv(r"C:\datascince\spam.csv",encoding='latin-1')
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
df.columns=['label','message']
df
```

```python
#Encode the labels:ham->0,spm->1
df['label']=df['label'].map({'ham':0,'spam':1})
df
```

```python
#Extract features and labels
X=df['message']
Y=df['label']
```

```python
#Convert theb text data into numerical data
cv=CountVectorizer()
X=cv.fit_transform(X)#fit and transform the data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
#train the navie Bayes Classifier
clf=MultinomialNB()
clf.fit(X_train,Y_train)
```

```python
#Evaluate the model
Y_pred=clf.predict(X_test)
print(f'Accuracy:{accuracy_score(Y_test,Y_pred)}')
```

```python
#Save the trained model and CountVectorizer for later use
with open('nlp_model.pkl','wb')as model_file:
    pickle.dump(clf,model_file)
with open('tranform.pkl','wb')as cv_file:
    pickle.dump(cv,cv_file)
print(f"model saved.")
```
